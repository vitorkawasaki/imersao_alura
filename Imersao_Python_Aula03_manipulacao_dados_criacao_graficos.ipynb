##1.6 - Verificação das colunas DF

```python
df_principal.columns
```

##1.7 - Copia do dataframe com as colunas escolhidas

```python
df_principal = df_principal[['Ativo', 'Data', 'Último (R$)', 'Var. Dia (%)']].copy()
df_principal
```

###1.7.1 - Renomeando colunas

* .copy() - faz uma cópia do original

```python
df_principal = df_principal.rename(columns={'Ativo': 'ativo',
                                            'Data':'data',
                                            'Último (R$)':'valor_final',
                                            'Var. Dia (%)':'var_dia'
                                            }).copy()
# df_principal recebe df_principal e renomeia as colunas...
df_principal
```

##1.8 - Criando novas colunas

* Análises do DataFrame

```python
df_principal['var_pct'] = df_principal['var_dia'] / 100
# cria uma nova coluna chamada 'var_pcte' recebe df_principal['var_dia] dividido por 100
df_principal
```

```python
df_principal['valor_inicial'] = df_principal['valor_final'] / (df_principal['var_pct']+1)
df_principal
```

##1.9 Juntando duas abas da mesma coluna - Merge

```python
df_principal = df_principal.merge(df_total_acoes, left_on='ativo',
                                  right_on='Código', how='left')
# df_principal recebe df_principal e junta: df_total_acoes. left é o principal e right o total_acoes,
# o how indica qual é o principal.
df_principal
```

###1.9.1 - Verificação das colunas DF

```python
df_principal.columns
```

###1.9.2 - Renomeando colunas

```python
df_principal = df_principal.rename(columns={'Código':'codigo',
                                            'Qtde. Teórica':'qtde_teorica'
                                            }).copy()
# df_principal recebe df_principal e renomeia as colunas...
df_principal
```

##1.10 - Eliminando coluna

```python
df_principal = df_principal.drop(columns=['codigo'])
# df_principal recebe df_principal e elimina a coluna codigo
df_principal
```

##1.11 - Criando coluna variação R$

```python
df_principal['variacao_rs'] = (df_principal['valor_final'] - df_principal['valor_inicial']) * df_principal['qtde_teorica']
df_principal
```

###1.11.1 - Corrigindo valores quebrados da coluna variacao_rs

```python
pd.options.display.float_format = '{:.2f}'.format
```

```python
df_principal
```

###1.11.2 - Mudando o tipo da coluna qtde_teorica

```python
df_principal['qtde_teorica'] = df_principal['qtde_teorica'].astype(int)
df_principal
```

##1.12 - Criando a coluna usando o ChatGPT

```python
df_principal['resultado'] = df_principal['variacao_rs'].apply(lambda x: 'Subiu' if x > 0 else ('Desceu' if x < 0 else 'Estavel'))
df_principal.sample(5)
# sample serve para pedir de forma randômica verificação dos dados e dentro do (), a quantidade de linhas que deseja
```

# Análises

```python
# Calculando o maior valor
maior = df_principal['variacao_rs'].max()

# Calculando o menor valor
menor = df_principal['variacao_rs'].min()

# Calculando a média
media = df_principal['variacao_rs'].mean()

# Calculando a média de quem subiu
media_subiu = df_principal[df_principal['resultado'] == 'Subiu']['variacao_rs'].mean()

# Calculando a média de quem desceu
media_desceu = df_principal[df_principal['resultado'] == 'Desceu']['variacao_rs'].mean()

# Exibindo os resultados
print("Maior:\t\tR$", '{:,.2f}'.format(maior))
print("Menor:\t\tR$", '{:,.2f}'.format(menor))
print("Média:\t\tR$", '{:,.2f}'.format(media))
print("Média de quem subiu:\tR$", '{:,.2f}'.format(media_subiu))
print("Média de quem desceu:\tR$", '{:,.2f}'.format(media_desceu))
```

#2 - Separando apenas o que Subiu

```python
df_principal_subiu = df_principal[df_principal['resultado'] == 'Subiu']
df_principal_subiu
```

##2.1 - Agrupando e somando por segmento

```python
df_analise_segmento = df_principal_subiu.groupby('segmento')['variacao_rs'].sum()
df_analise_segmento
```

```python
df_analise_saldo = df_principal.groupby('resultado')['variacao_rs'].sum().reset_index()
df_analise_saldo
```

```python
fig = px.bar(df_analise_saldo, x='resultado', y='variacao_rs', text='variacao_rs', title='Variação Reais por Resultado')
fig.show()
```

```python
# Arredondando os valores da coluna 'variacao_rs' para duas casas decimais
df_analise_saldo['variacao_rs'] = df_analise_saldo['variacao_rs'].round(2)

# Criando o gráfico de barras com os valores arredondados
fig = px.bar(df_analise_saldo, x='resultado', y='variacao_rs', text='variacao_rs', title='Variação Reais por Resultado')
fig.show()
```

```python
# Arredondando os valores da coluna 'variacao_rs' para duas casas decimais
df_principal['variacao_rs'] = df_principal['variacao_rs'].round(2)

# Criando o gráfico de pizza
fig = px.pie(df_principal, values='variacao_rs', names='segmento', title='Variação Reais por Segmento')
fig.update_traces(textposition='inside', textinfo='percent+label')  # Mostra o percentual e o rótulo dentro das fatias
fig.show()
```

```python
# Criando o gráfico de barras
fig = px.bar(df_principal, x='categoria', y='variacao_rs', title='Variação Reais por Categoria')

# Atualizando o layout do gráfico para exibir os números com duas casas decimais
fig.update_layout(yaxis_tickformat=".2f")

# Exibindo o gráfico
fig.show()
```

```python
# Agrupando os dados pelo segmento e contando o número de empresas em cada grupo
df_contagem_empresas = df_principal.groupby('categoria').size().reset_index(name='quantidade_empresas')

# Criando o gráfico de barras horizontais
fig = px.bar(df_contagem_empresas, y='categoria', x='quantidade_empresas', orientation='h',
             title='Quantidade de Empresas por Categoria')

# Exibindo o gráfico
fig.show()
```
